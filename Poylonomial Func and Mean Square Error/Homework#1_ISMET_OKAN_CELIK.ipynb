```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')
noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 7 * x + 15 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
```

```python
x=[row[0] for row in x]
y=[row[0] for row in y]
def polyfunc(M):
    curve=np.polyfit(x,y,M)
    poly=np.poly1d(curve)
    new_y=[]
    for i in x:
        y_update=poly(i)
        new_y.append(y_update)
    return new_y
```

```python
def plotting_func(M):
    ax=plt.subplot()
    ax.plot(x,y,'ro',label='Data')
    ax.plot(x,polyfunc(M),'o',label='Polygon')
    plt.title('M =' + str(M))
    plt.xlabel('x Values')
    plt.ylabel('y Values')
    plt.legend()
    plt.show()
```

```python
plotting_func(2)
```

```python
plotting_func(3)
```

```python
plotting_func(4)
```

```python
plotting_func(5)
```

```python
plotting_func(6)
```

```python
#Calculation Function of Mean Square Error
def MSE(y,y_new):
    y_list=[]
    for i in range(len(y)):
        each_y=(y[i]-y_new[i])**2
        y_list.append(each_y)
    E=(1/(2*len(y)))*sum(y_list)
    return E
```

```python
MSE_list=[]
for m in range(1,9):
    y_new=polyfunc(m)
    MSE_list.append(MSE(y,y_new))
print(MSE_list)
```

```python
#Plotting of Mean Square Error

m=[i for i in range(1,9)]
                     
plt.plot(m,MSE_list,'-o')
plt.xlabel('m values')
plt.ylabel('Mean Square Errors')
plt.show()
```

### After examining the graph above we can say that m=8 is the best m because it gives us minimum error.

```python
noise_list=[150,200,400,600,1000]
def New_Noisy_Data(noise_scale):
    number_of_samples = 50
    x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
    y = 7 * x + 15 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1) 
    x=[row[0] for row in x]
    y=[row[0] for row in y]
    return x,y
for noise_scale in noise_list:
    New_Data=New_Noisy_Data(noise_scale)
    x=New_Data[0]
    y=New_Data[1]
    plotting_func(8)
```

### As we can see on the graphs, it is hard to fit an accurate polynomial as a model when the noise level increases. Noise values are respectively increased from 150 to 1000, and on each level, it is become more challenging to see any pattern to make sense of the data. In this situation system less likely to perform good. Because data is not accurate it is too noisy.
