```python
family_list=timeseries.columns[1:]

for i in family_list:
    fig,ax=plt.subplots()
    ax.plot(timeseries['date'],timeseries[i],label=i)
    ax.plot(oil_df['date'],oil_df['dcoilwtico'],label='oil_price')
    ax.set_title(i)
    ax.set_xlabel('Date')
    ax.set_ylabel('Price')
    plt.legend()
    plt.show
```

```python
list_of_dates=oil_df['date'].values

correalation_oil_families=[]
for a in family_list:
    index_list=[]
    for i in list_of_dates:
        index_num=timeseries[timeseries['date'] ==i].index.values
        index_list.append(index_num)
    index_flatten_matrix = [val for sublist in index_list for val in sublist]

    family=[]
    for i in index_flatten_matrix:
        family.append(timeseries.loc[i,a])

    deneme=pd.DataFrame()
    deneme1=family
    deneme2=oil_df['dcoilwtico'].loc[1:].values[:len(family)]
    deneme[1]=deneme1
    deneme[2]=deneme2
    correalation_oil_families.append(deneme.corr().loc[1][2])
```
